```python
from PIL import Image
import os
```

```python

# Load the image
image_path = "images.png"
img = Image.open(image_path)

# Get image dimensions
width, height = img.size
print(f"Original image dimensions: {width} x {height}")

# Calculate the width of each section (divide horizontally)
section_width = width // 6
print(f"Each section width: {section_width} pixels")
```

```python

# Extract and save each section in the current working directory
for i in range(6):
    # Calculate the left and right boundaries for this section
    left = i * section_width
    right = (i + 1) * section_width
    
    # Extract the section (crop horizontally)
    section = img.crop((left, 0, right, height))
    
    # Save the section in the current working directory
    output_path = f"image_{i+1}.png"
    section.save(output_path)
    print(f"Saved: {output_path} (dimensions: {section_width} x {height})")

print(f"\nAll 6 images have been saved to the current working directory!")
```
